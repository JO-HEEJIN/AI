```python
# drop the 0,1,2 rows because of starting page
df = pd.read_excel(xlsx_file_path, header = None)
```

```python
mask = df.iloc[:, [1,2,3,4,5,6,7,11]].isnull().all(axis=1) & df.iloc[:, [8,9]].notnull().any(axis=1)



# df.insert(12, 'test',mask)
# df.insert(13, "test2", mask ) 
# df.insert(14, "test3", mask ) 




# 8열에서 조건에 맞는 경우 새로운 열에 저장하고 그 열의 값을 모두 한 칸씩 올려준다
df["new"] = df.loc[mask, [8]]
df["new"] = df["new"].shift(-1)

# 9열에서 조건에 맞는 경우 새로운 열에 저장하고 그 열의 값을 모두 한 칸씩 올려준다
df["new2"] = df.loc[mask, [9]]
df["new2"] = df["new2"].shift(-1)

# mask가 true인 행은 모두 drop (왜냐하면 이미 mask가 true인 값들을 윗 행으로 올려서)
df = df.drop(df[mask].index)
```

```python
df
```

```python
# new 와 8열을 합친다.
df[8] = np.where(pd.isnull(df[8]), df['new'], df[8])
# new2와 9열을 합친다.
df[9] = np.where(pd.isnull(df[9]), df['new2'], df[9])
df
```

```python
df = df.drop(["new","new2"], axis=1)
```

##### set the columns values

```python
columns_to_fill ="gov,dept,policy,danwi,task,acct1,acct2".split(",") # '부서', '정책', '단위', '세부사업', '편성목', '편성운영비'
# columns_to_fill = header[:5] 이 방법으로도 가능

for col in columns_to_fill:
    df[col] = df[col].fillna(method='ffill')
```

```python
df
```

fill df['bizdetail']

```python
df['bizdetailadd'] = df['bizdetailadd'].astype(str).replace('nan', '')
df['detailamt'] = df['detailamt'].astype(str).replace('nan', '')
df['bizdetail'] = df['bizdetailadd'] + df['detailamt']
```

if df['bizdetailadd'] startwith('1)') then delete the cell's content

```python
# Create a boolean mask where True indicates that the cell value starts with '1)'
mask = df['bizdetailadd'].str.startswith('O')

# Set the values of the selected cells to an empty string
df.loc[mask, 'bizdetailadd'] = ''
```

```python
# 1. DataFrame after dropping empty rows:
df = df[df['bizdetail'] != '']
```
